# forma moderna de hacer las tecnicas de limpieza de datos con Scikit_learn


en este archivo conoceremos las formas modernas de realizar las tecnicas de limpieza de datos tanto **eliminación como imputación**

---

```python
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer

```

## carga del data set

```python
data = {
    'edad': [25, 30, np.nan, 45, 50, np.nan,35],
    'salario': [30000, 45000, 50000, np.nan,60000, 55000, np.nan],
    'ciudad': ['madrid','barcelona',None, 'valencia', 'madrid','barcelona','valencia'],
    'experiencia_en_años': [2,5,7, np.nan, 10,8,4]
}
df = pd.DataFrame(data)
print("data set oringinal")
print(df)
print("\n"+"="*50+ "\n")
```

## separación de columnas

separamos las columnas numericas del data frame 

```python
columnas_numericas = ['edad', 'salario','experiencia_en_años']

## creacion de imputador con estrategia 'mean' (media)
imputer_media = SimpleImputer(strategy = 'mean') ## que hace simpleImputer
# ajustar y transformar
df_imputado = df.copy()
df_imputado[columnas_numericas] = imputer_media.fit_transform(df[columnas_numericas])

print ("dataset con SimpleImputer (media)")
print (df_imputado)
```
