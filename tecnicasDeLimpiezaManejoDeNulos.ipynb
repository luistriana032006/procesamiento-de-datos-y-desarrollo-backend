# creación de un data set de ejemplo con valores nulos

acá tenemos un diccionario que contiene 4 columnas con 7 filas osea una tabla cada columna se representa como una lista y cada valor de la columna contiene un valor especifico para esos registros 

```python
data = {
    'edad': [25, 30, np.nan, 45, 50, np.nan,35],
    'salario': [30000, 45000, 50000, np.nan,60000, 55000, np.nan],
    'ciudad': ['madrid','barcelona',None, 'valencia', 'madrid','barcelona','valencia'],
    'experiencia_en_años': [2,5,7, np.nan, 10,8,4]
}
df = pd.DataFrame(data)
print("data set oringinal")
print(df)
print("\n"+"="*50+ "\n")
```

## verificación de valores nulos

queremos mirar si existen valores nulos en nuestro data set
pero como lo hacemos: 

**vamos a usar la funcion ``isnull(())``** que nos devuelve TRUE en donde hay nulos


```python
print("hay valores nulos ??")
print(df.isnull()) ## devuelve true si tenemos valores nulos
print("\n"+"="*50 + "\n")
```
